```python
import pandas as pd
```

```python
yagtf = pd.read_csv("../Yaloifolia_all.gtf_corrected.gtf",sep="\t",header=None,comment="#")
yagtf.head()
```

```python
yfgtf = pd.read_csv("../YfilamentosavarC3HAP1v3.1.gene.gtf_corrected.gtf",sep="\t",header=None,comment="#")
yfgtf.head()
```

```python
# rename all chromosomes as for the genome file
mapya = {i:f"Ya_{i}" for i in list(yagtf[0].unique())}
mapyf = {i:f"Yf_{i}" for i in list(yfgtf[0].unique())}
```

```python
mapya
```

```python
yagtf[0] = yagtf[0].map(mapya)
yagtf.head()
```

```python
yfgtf[0] = yfgtf[0].map(mapyf)
```

```python
# stick the two dataframes together & write updated gtf file
yggtf = pd.concat([yagtf,yfgtf])
len(yggtf.index)==len(yagtf.index)+len(yfgtf.index)
```

```python
yggtf.to_csv("./yg_insilico_corrected.gtf",sep="\t",header=False,index=False)
```
